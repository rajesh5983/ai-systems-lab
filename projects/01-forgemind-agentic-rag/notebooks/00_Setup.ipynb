```python
# Install required packages
import subprocess
import sys

packages = [
    'openai',
    'langchain', 
    'langflow',
    'langchain-community',
    'langchain-openai',
    'chromadb',
    'python-dotenv',
    'requests'
]

print("Installing required packages...")
for package in packages:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package])
print("✓ All packages installed successfully!")
```

```python
# Configure API Keys
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Verify API keys are configured
openai_key = os.getenv('OPENAI_API_KEY')

print("Environment Configuration Check:")
print(f"✓ OpenAI API Key configured: {bool(openai_key)}")

if not openai_key:
    print("\n⚠️  Warning: OpenAI API key is not configured!")
    print("Please create a .env file with your API key or set environment variables.")
else:
    print("\n✅ OpenAI API key is configured. Ready to start the course!")
```
